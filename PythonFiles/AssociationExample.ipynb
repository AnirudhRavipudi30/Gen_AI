```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns

# Simulating stock market data: Returns of Microsoft (MSFT) and NASDAQ-100 Index (NDX)
np.random.seed(42)

# Generate 252 daily returns (1 year of trading days)
nasdaq_returns = np.random.normal(loc=0.002, scale=0.015, size=252)  # NASDAQ-100 returns
msft_returns = 0.85 * nasdaq_returns + np.random.normal(loc=0.001, scale=0.012, size=252)  # Microsoft returns

# Creating a DataFrame
stock_comparison_data = pd.DataFrame({
    "NASDAQ-100 Returns": nasdaq_returns,
    "Microsoft Returns": msft_returns
})

# Calculate Pearson Correlation
corr_coefficient_msft, p_value_msft = stats.pearsonr(
    stock_comparison_data["NASDAQ-100 Returns"], stock_comparison_data["Microsoft Returns"]
)

# Scatter Plot with Regression Line
plt.figure(figsize=(8, 5))
sns.regplot(
    x=stock_comparison_data["NASDAQ-100 Returns"],
    y=stock_comparison_data["Microsoft Returns"],
    scatter_kws={"alpha": 0.5}
)
plt.xlabel("NASDAQ-100 Returns")
plt.ylabel("Microsoft Returns")
plt.title(f"Association between NASDAQ-100 and Microsoft Returns\n"
          f"Pearson Correlation: {corr_coefficient_msft:.2f}, P-Value: {p_value_msft:.4f}")
plt.grid(True)
plt.show()

# Print results
print(f"Pearson Correlation Coefficient: {corr_coefficient_msft:.4f}")
print(f"P-Value: {p_value_msft:.50f}")  # Display full decimal format

# Conclusion Interpretation
if p_value_msft < 0.05:
    print("Conclusion: Strong association - Microsoft and NASDAQ-100 returns move together.")
else:
    print("Conclusion: No significant association found.")
```
