```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Simulating stock data to demonstrate Simpson's Paradox in stock returns

# Generate synthetic stock returns for 2 years (8 quarters)
np.random.seed(42)
quarters = ["Q1-Y1", "Q2-Y1", "Q3-Y1", "Q4-Y1", "Q1-Y2", "Q2-Y2", "Q3-Y2", "Q4-Y2"]
market_condition = ["Bull", "Bear", "Bull", "Bear", "Bull", "Bear", "Bull", "Bear"]  # Alternating bull and bear markets

# Stock returns in bull markets are positive, in bear markets are negative
returns_year_1 = [10, -5, 8, -4]  # Tesla's returns in Y1
returns_year_2 = [12, -3, 9, -2]  # Tesla's returns in Y2

# Create DataFrame
simpson_stock_data = pd.DataFrame({
    "Quarter": quarters,
    "Market Condition": market_condition,
    "Year": ["Year 1"]*4 + ["Year 2"]*4,
    "Stock Return (%)": returns_year_1 + returns_year_2
})

# Compute overall yearly returns (misleading aggregated data)
overall_returns = simpson_stock_data.groupby("Year")["Stock Return (%)"].mean()

# Scatter plot to visualize stock performance in different markets
plt.figure(figsize=(10,5))
sns.barplot(x="Quarter", y="Stock Return (%)", hue="Market Condition", data=simpson_stock_data)
plt.axhline(y=0, color='black', linestyle='dashed')
plt.title("Stock Returns in Bull vs Bear Markets (Simpson's Paradox Example)")
plt.ylabel("Stock Return (%)")
plt.xlabel("Quarter")
plt.legend(title="Market Condition")
plt.xticks(rotation=45)
plt.show()

# Print Overall Returns
print("\n📌 Overall Yearly Returns (Aggregated Data):")
print(overall_returns)
```
